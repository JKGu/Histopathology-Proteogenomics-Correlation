## Import libraries

```python
from scipy.stats import spearmanr
import pandas as pd
from scipy.io import loadmat
import numpy as np
```

## Read in mRNA data

```python
rna_raw_data = 'OV__geneExp.txt'
rna_df = pd.read_csv(rna_raw_data,sep='\t', engine='python')
rna_df
```

## Read in proteinomics data

```python
protein_raw_data = 'OV__protein_iTRAQ_allPeptides.txt'
protein_df = pd.read_csv(protein_raw_data,sep='\t', engine='python')
protein_df
```

## Read in tissue slide image data

```python
image_raw_data = 'imData.csv'
image_df = pd.read_csv(image_raw_data)
image_df
```

## Data cleaning and preprocessing

```python
for x in rna_df.columns:
    if x.startswith('TCGA'):
        rna_df.rename(columns = {x:x [0:12]}, inplace = True) 
rna_df.columns
```

```python
rna_df
```

```python
rna_df = rna_df.loc[:,~rna_df.columns.duplicated()]
rna_df
```

```python
rna_df = rna_df[rna_df.GeneSymbol != '?']
rna_df
```

```python
rna_df = rna_df.drop_duplicates(subset='GeneSymbol')
rna_df
```

```python
rna_df.set_index('GeneSymbol', inplace=True)
rna_df
```

```python
rna_df = rna_df.drop('EntrezID',1)
rna_df
```

```python
protein_df = protein_df.dropna(axis=0)
```

```python
for x in protein_df.columns:
    if x.startswith('TCGA'):
        protein_df.rename(columns = {x:x [0:12]}, inplace = True) 
protein_df = protein_df.loc[:,~protein_df.columns.duplicated()]
protein_df
```

```python
protein_df = protein_df.drop_duplicates(subset='GeneSymbol')
protein_df
```

```python
protein_df.set_index('GeneSymbol', inplace=True)
protein_df
```

```python
protein_df = protein_df.drop('Description',1)
protein_df = protein_df.drop('Organism',1)
protein_df = protein_df.drop('Chromosome',1)
protein_df = protein_df.drop('Locus',1)
protein_df
```

```python
image_df = image_df.transpose()
image_df
```

```python
new_header = image_df.iloc[0]
image_df = image_df[1:] 
image_df.columns = new_header 
image_df
```

```python
for x in image_df.columns:
    if x.startswith('TCGA'):
        image_df.rename(columns = {x:x [0:12]}, inplace = True) 
image_df
```

```python
matched_id = rna_df.columns.intersection(protein_df.columns.intersection(image_df.columns))
matched_id
```

```python
matched_gene = rna_df.index.intersection(protein_df.index)
matched_gene
```

```python
rna_df=rna_df[matched_id]
rna_df=rna_df.loc[matched_gene]
rna_df
```

```python
protein_df=protein_df[matched_id]
protein_df=protein_df.loc[matched_gene]
protein_df
```

```python
image_df=image_df[matched_id]
image_df
```

## Data analysis

```python
cor_rna_df = pd.DataFrame(np.zeros((image_df.shape[0], rna_df.shape[0])))
cor_rna_df.columns = rna_df.index
cor_rna_df.index = image_df.index
cor_rna_df.columns.name = 'mRNAGeneSymbol'
cor_rna_df.index.name = 'imageFeatures'
cor_rna_df
```

```python
cor_protein_df = pd.DataFrame(np.zeros((image_df.shape[0], protein_df.shape[0])))
cor_protein_df.columns = protein_df.index
cor_protein_df.index = image_df.index
cor_protein_df.columns.name = 'proteinGeneSymbol'
cor_protein_df.index.name = 'imageFeatures'
cor_protein_df
```

```python
counter = 0
total = image_df.shape[0] * protein_df.shape[0]
percent = round(total/100)
if percent <= 0:
    percent = 1
print(f'Correlation analysis with {total} pairs...')
for i in range(image_df.shape[0]):
    l1 = image_df.iloc[i].tolist()
    for j in range(rna_df.shape[0]):
        l2 = rna_df.iloc[j].tolist()
        l3 = protein_df.iloc[j].tolist()
        counter += 1
        cor_rna_df.iat[i,j] = spearmanr(l1, l2)[1]
        cor_protein_df.iat[i,j] = spearmanr(l1, l3)[1]
        if counter % percent == 0:
            print(f'Processed {round(counter/total*100)}%...')
cor_rna_df.to_csv('rna_image_cor.csv')  
cor_protein_df.to_csv('protein_image_cor.csv')  
```

```python
cor_df = pd.DataFrame(np.zeros((3, 3)))
cor_df.columns = ['Positive Correlation','Noncorrelation','Negative Correlation']
cor_df.index = ['Positive Correlation','Noncorrelation','Negative Correlation']
cor_df.columns.name = 'image-protein'
cor_df.index.name = 'image-mRNA'
cor_df
```

```python
for i in range(cor_rna_df.shape[0]):
    for j in range(cor_rna_df.shape[1]):
        pvalue_r = cor_rna_df.iat[i,j]
        pvalue_p = cor_protein_df.iat[i,j]
        x = ''
        y = ''
        if pvalue_r > 0.3:
            x = 'Positive Correlation'
            if pvalue_p > 0.3:
                y = 'Positive Correlation'
            elif pvalue_p < -0.3:
                y = 'Negative Correlation'
            elif pvalue_p <= 0.3:
                y = 'Noncorrelation'
        elif pvalue_r < -0.3:
            x = 'Negative Correlation'
            if pvalue_p > 0.3:
                y = 'Positive Correlation'
            elif pvalue_p < -0.3:
                y = 'Negative Correlation'
            elif pvalue_p <= 0.3:
                y = 'Noncorrelation'
        elif pvalue_r <= 0.3:
            x = 'Noncorrelation'
            if pvalue_p > 0.3:
                y = 'Positive Correlation'
            elif pvalue_p < -0.3:
                y = 'Negative Correlation'
            elif pvalue_p <= 0.3:
                y = 'Noncorrelation'     
        if x != '' and y != '':
            cor_df.at[x,y]+=1 
```
